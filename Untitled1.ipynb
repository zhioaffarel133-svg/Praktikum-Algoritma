Double Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoubleLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" <-> ")
            temp = temp.next
        print("None")

    def insert_middle(self, prev_node_data, new_data):
        temp = self.head
        while temp:
            if temp.data == prev_node_data:
                new_node = Node(new_data)
                new_node.next = temp.next
                new_node.prev = temp
                if temp.next:
                    temp.next.prev = new_node
                temp.next = new_node
                return
            temp = temp.next
        print("Node acuan tidak ditemukan")

    def delete_by_key(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                if temp.prev:
                    temp.prev.next = temp.next
                else:
                    self.head = temp.next
                if temp.next:
                    temp.next.prev = temp.prev
                temp = None
                return
            temp = temp.next

dll = DoubleLinkedList()
n1 = Node("Satu")
n2 = Node("Dua")
n3 = Node("Tiga")

dll.head = n1
n1.next = n2
n2.prev = n1
n2.next = n3
n3.prev = n2

print("=== Data Awal ===")
dll.print_list()

dll.insert_middle("Dua", "Sisipan")
print("\n=== Setelah Penyisipan di Tengah ===")
dll.print_list()

dll.delete_by_key("Dua")
print("\n=== Setelah Dihapus dengan Kata Kunci 'Dua' ===")
dll.print_list()
```

Circular Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        temp = self.head
        if self.head is not None:
            while True:
                print(temp.data, end=" -> ")
                temp = temp.next
                if temp == self.head:
                    break
        print("(kembali ke head)")

    def insert_end(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

    def delete_end(self):
        if self.head is None:
            return
        if self.head.next == self.head:
            self.head = None
            return
        temp = self.head
        while temp.next.next != self.head:
            temp = temp.next
        temp.next = self.head

cll = CircularLinkedList()
n1 = Node("Januari")
n2 = Node("Februari")
n3 = Node("Maret")

cll.head = n1
n1.next = n2
n2.next = n3
n3.next = cll.head

print("=== Data Awal ===")
cll.print_list()

cll.insert_end("April")
print("\n=== Setelah Penyisipan di Akhir ===")
cll.print_list()

cll.delete_end()
print("\n=== Setelah Penghapusan di Akhir ===")
cll.print_list()
```
